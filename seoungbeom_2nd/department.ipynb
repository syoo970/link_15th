```python
import pandas as pd
import numpy as np

sales = pd.read_csv('C:\\Users\\syoo9\\백화점매출예측\\DEPT\\백화점_하루매출.csv', sep=",")
sales = pd.DataFrame(sales.values[:,-1] , columns = ['sales'])
arr_sales = np.asarray(sales)

sales.head()
```

```python
import pandas as pd 
import numpy as np

all_pur = [] 

for x in range(0,50): #50 *10 = 500, 데이터 부족으로 인해서
    for i in range(1,11):
        filename = "C:\\Users\\syoo9\\백화점매출예측\\DEPT\\백화점_구매지수_" + str(i) + ".csv"    
        pur = pd.read_csv(filename, sep=",")
        pur = pd.DataFrame(pur.values[:,1:])
        arr_pur = np.asarray(pur)
        #print(arr_pur.ndim) #차원 확인
        arr_pur = arr_pur.reshape(300,200,1) 
        #print('reshape후 차원 ',arr_pur.ndim)
        #arr_pur.shape
        #pur.head()
        all_pur.append(arr_pur) 
        print(filename) 
    
all_pur_arr = np.asarray(all_pur)
```

```python
cnt = 1
sales_train = [] 
sales_test = [] 
sales_valid = [] 

for i in arr_sales:
       
    if cnt <= 300: #train 데이터
        sales_train.append(i)
    elif cnt > 300 and cnt <= 400: #validation 데이터
        sales_valid.append(i)
    else: #test 데이터
        sales_test.append(i)
        
    print(cnt) 
    cnt += 1

arr_sales_train = np.asarray(sales_train)
arr_sales_test = np.asarray(sales_test)
arr_sales_valid = np.asarray(sales_valid)
```

```python
arr_sales_train.shape , arr_sales_test.shape , arr_sales_valid.shape
```

```python
#구매지수 데이터!! , input임
cnt = 1
pur_train = [] 
pur_test = [] 
pur_valid = [] 

for i in all_pur_arr:
       
    if cnt <= 300: #train 데이터 300개
        pur_train.append(i)
    elif cnt > 300 and cnt <= 400: #valid 데이터 100개
        pur_valid.append(i)
    else: #test 데이터 100개
        pur_test.append(i)
        
    print(cnt) 
    cnt += 1

arr_pur_train = np.asarray(pur_train)
arr_pur_test = np.asarray(pur_test)
arr_pur_valid = np.asarray(pur_valid)
```

```python
#arr_pur shape 찍어서 확인, 300 100 100 으로 나누어지는것 확인
arr_pur_train.shape , arr_pur_test.shape , arr_pur_valid.shape
```

```python
# 패키지 불러오기
import numpy as np
from keras.models import Sequential
from keras.layers import Dense , Dropout , Flatten
from keras.layers import Conv2D , MaxPooling2D

#이미지의 크기 세팅 
width = 300
height = 200

x_train, y_train = arr_pur_train , arr_sales_train
x_val, y_val = arr_pur_valid , arr_sales_valid
x_test, y_test = arr_pur_test , arr_sales_test

# 모델 구성하기
model = Sequential()
model.add(Conv2D(32,(3,3) , activation='relu' , input_shape=(width,height,1))) 
model.add(MaxPooling2D(pool_size=(2,2))) 
model.add(Conv2D(32,(3,3) , activation='relu')) 
model.add(MaxPooling2D(pool_size=(2,2))) 
model.add(Flatten()) 
model.add(Dense(256 , activation='relu')) 
model.add(Dense(1)) 

# 컴파일
model.compile(loss='mse', optimizer='adam')

# 학습 
hist = model.fit(x_train, y_train, batch_size=64, epochs=30, validation_data=(x_val, y_val))
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

plt.plot(hist.history['loss'])
plt.plot(hist.history['val_loss'])
plt.ylim(0.0, 300.0)
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'val'], loc='upper left')
plt.show()

#모델 평가하기
score = model.evaluate(x_test, y_test, batch_size=32)

print(score)

yhat_test = model.predict(x_test, batch_size=32)

%matplotlib inline
import matplotlib.pyplot as plt

plt_row = 7
plt_col = 7

plt.rcParams["figure.figsize"] = (10,10)

f, axarr = plt.subplots(plt_row, plt_col)

for i in range(plt_row*plt_col):
    sub_plt = axarr[(int)(i/plt_row), i%plt_col]
    sub_plt.axis('off')
    sub_plt.imshow(x_test[i].reshape(width, height))
    sub_plt.set_title('R %d P %.1f' % (y_test[i][0], yhat_test[i][0]))

plt.show()
```
